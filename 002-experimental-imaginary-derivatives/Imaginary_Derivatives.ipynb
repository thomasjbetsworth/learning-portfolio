The standard derivative of a real-valued function $f(x)$ is defined as follows:

\begin{equation}
f'(x) = \lim_{h \to 0}{\frac{f(x+h)-f(x)}{h}}
\end{equation}

Other approaches, however, opt to define $h$ as an infinitesimal rather than using a limit, as limits are often hard to deal with both mathematically and computationally. In the latter case, many opt for minute finite differences, rather than true infinitesimals.

One way of viewing infinitesimals is as contents of the real number $0$. In other words, as well as being a real number, $0$ also acts as a range or set of infinitesimal values. Each of these infinitesimals may be represented by a real number multiplied by an infinitesimal unit, $\omicron$. For all real $r$:

\begin{equation}
r\omicron ∈ 0
\end{equation}

This is a formalisation of the concept of $0$ containing infinitesimals, in set form. This can also be generalised beyond $0$, since infinitesimals can be added to any real number. For all real $r$ and $n$:

\begin{equation}
n' + r\omicron ∈ n
\end{equation}

Thus, every real number may also be considered a set of infinitesimals. The prime sign signifies that the number which is added to the infinitesimal is not itself a range of infinitesimals, but a particular value within said range, where there is no infinitesimal part. The distinction is best explained by imagining the number $n$ plotted on the real number line, and a second line above, representing the infinitesimal contents of $n$. $n$ describes the entire upper line, while $n'$ describes only the point on said line where the infinitesimal term $r\omicron$ has a coefficient $r=0$. $n'$ may be said to have a higher precision than $n$. We may define a rounding function $R(n'+r\omicron)=n$ which takes in a value of infinitesimal precision and returns the real number which contains it.

Based on this system, it is possible to differentiate real-valued functions using a derivative defined as follows:

\begin{equation}
f'(x) = R\left(\frac{f(x'+\omicron)-(f(x))'}{\omicron}\right)
\end{equation}

Note: $f(x)'$ is not to be confused with $f'(x)$. The latter is the derivative; the former refers to the precision level discussed earlier.

Now we shall explore a different way of viewing infinitesimals. Since infinitesimals are defined as the elements of the number zero, and numbers with a real and infinitesimal term (a.k.a. numbers of infinitesimal precision) as elements of real numbers, we may display every such number on a two-dimensional plot, where the axes represent the coefficients of the real and infinitesimal components respectively. For clarity, we shall assume the real components to take the horizontal, and the infinitesimal ones to take the vertical. Here the distinction can readily be drawn between $n$ and $n'$: $n'$ is a value of infinitesimal precision at coordinates $(n,0)$, while $n$ is represented by a vertical line which intersects the real axis at point $n$. The fact that said line passes through all infinitesimal values indicates the previously stated fact that $n$ contains $n'+r\omicron$ for all real $r$; the value of $n$ and thus $n'$ fixes the real term (horizontal coordinate), but the infinitesimal term (vertical coordinate) is free to vary, resulting in a vertical line.

Imagining the infinitesimal terms as orthagonal to the real ones leads to one key mathematical feature: $n'+r_1o$ cannot be said to be greater than or less than $n'+r_2o$, for distinct $r_1$ and $r_2$, since the $<$ and $>$ operations are only applicable to differences in a single dimension: the real value.

Upon retrospection, it may seem curious that the derivative formula introduced above is not invalidated by this altered criterion. It is not intuitively obvious that a normalised shift in an orthagonal dimension could be used in place of the standard limit technique. It is this ability to utilise orthagonal dimensions, however, which allows the infinitesimal system to be replaced by an equivalent which utilises an existing mathematical construct: the complex number. Instead of making an infinitesimal change of $\omicron$, we make a change in the imaginary dimension, like so:

\begin{equation}
f'(x) = \text{Re}\left(\frac{f(x+i)-f(x)}{i}\right)
\end{equation}

Here, instead of the rounding function $R$ for infinitesimals, the function $\text{Re}$ isolates the real component of the inner formula.

Counter-intuitively, both the infinitesimal and imaginary approach find perfect derivatives of a function without utilising limits, by very fact of being orthagonal (throughout this writing, we are dealing only with real-valued $f$). In both cases, the constant coefficient, sum and chain rules follow naturally from simple mathematical rules.

However, in contrast with infinitesimals, the scope of imaginary derivatives is limited to linear and quadratic functions. The proof of applicability within this range is as follows, where $f(x)=ax^2+bx+c$; we expect a result of $2ax+b$:

\begin{equation}
f'(x) = \text{Re}\left(\frac{(a(x+i)^2+b(x+i)+c)-(ax^2+bx+c)}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(\frac{(a(x^2+2xi+i^2)+b(x+i)+c)-(ax^2+bx+c)}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(\frac{ax^2+2axi+ai^2+bx+bi+c-ax^2-bx-c}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(\frac{2axi+ai^2+bi}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(2ax+ai+b\right)
\end{equation}

\begin{equation}
= 2ax+b
\end{equation}

In a cubic case, however, a counter-example can be given, thus proving that imaginary differentiation is not applicable to all higher power monomial terms. For this example, only a cubic term is included, so $f(x) = ax^3$; we expect a result of $3ax^2$:

\begin{equation}
f'(x) = \text{Re}\left(\frac{a(x+i)^3-ax^3}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(\frac{ax^3+3ai^2x+3aix^2+ai^3-ax^3}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(\frac{3ai^2x+3aix^2+ai^3}{i}\right)
\end{equation}

\begin{equation}
= \text{Re}\left(3aix+3ax^2+ai^2\right)
\end{equation}

\begin{equation}
= \text{Re}\left(3aix+3ax^2-a\right)
\end{equation}

\begin{equation}
= 3ax^2 - a
\end{equation}

Thus, for $a \not = 0$ (i.e. when the function is cubic), an extra constant term is introduced, on account of a higher power of $i$ being real. Since this is only the case for even powers of $i$, it follows that the imaginary derivative of any polynomial containing odd powers of $x$ will contain such anomalous terms (excepting the case of $x^1$, when, as has been shown, the derivative term is not anomalous, but desirable). This is because an odd power of $x+i$ must by necessity contain at least one odd power of $i$ in its expansion. When, like all terms, it is divided by $i$ prior to having its imaginary component clipped by the $\text{Re}$ function, an even power, and thus a real number, results. Even an even power of $x+i$ would contain in its expansion at least one term in an odd power of $i$, and the only place where this is acceptable, given the special case of $x^1$, is in a quadratic.

The reason that infinitesimals as herein defined don't run into a similar limitation is that we have implicitly assumed the potential for higher powers of $\omicron$, the infinitesimal unit. In the quadratic case, such a higher power would be present but trivial, as it would be cancelled down to a standard infinitesimal by division by $\omicron$ before rounding by the $R$ function. In a cubic case, however, one would still be left with a term in $\omicron^2$ even after this cancellation. The rounding function, which clips all infinitesimals, would also remove any of this new type; on this front, it makes no difference, but in order to perform algebraic manipulation, we must have a consistent definition for all positive powers of $\omicron$.

One way to imagine the higher-order infinitesimals which higher powers of $\omicron$ allow is as the contents of infinitesimals of one order lower. Thus:

\begin{equation}
n'\omicron^{p-1} + r\omicron^p ∈ n\omicron^{p-1}
\end{equation}

One can also imagine adding an extra orthagonal dimension for each additional order of infinitesimal, though both explanatory methods break down somewhat when faced with fractional powers of $\omicron$.

The key difference between higher powers of $\omicron$ and $i$ is that the former do not follow a cycle of sorts by which certain higher-order infinitesimals equal real numbers. This cycle, however, is prevalent in $i$; only odd powers of $i$ are purely imaginary. Thus, as we have seen, imaginary derivatives are only valid for quadratics, while infinitesimal ones are more widely applicable. Generally, however, any variable $l$ representing an orthagonal dimension will suffice to take the place of $\omicron$, given two critical requirements:

1. $l^0$ must equal a universal unit (usually $1$)
2. $l$ is acyclic under exponentiation.

The former requirement is vital because, as we have seen, many desired terms may be present on the numerator with coefficients of $l$, and these must be cancelled by division. It is a property common to many mathematical objects, including complex and infinitesimal values, but not, importantly, to elementwise multiplication over vectors. Thus, despite in a way being the most obvious choice for orthagonality, vectors cannot be used for this practise.

Let us define a vector, $\vec{x}$, in $D_{\text{in}}$ dimensions, but let us append to it an additional dimension; the coordinate in said dimension is $0$. Let us also define a function $f(\vec{x})$, which outputs a vector of length $D_{\text{out}}$. We stipulate that this output vector also retains the appended final value of $0$, as well, though if it is non-zero in the input, it may be acted upon by the function. Finally, let us define $\vec{l_{\text{in}}}$ as a $D_{\text{in}}+1$ dimensional vector, with values of $0$ in the first $D_{\text{in}}$ dimensions and a value of $1$ in the last, and $\vec{l_{\text{out}}}$ as an equivalent vector for the output dimensions. These will together act as our orthagonal unit, as $\omicron$ and $i$ did before. The reason there must be two of them is that the function has the potential to output a vector of a different length from its input. We may create a formula similar to earlier forms:

\begin{equation}
f'(\vec{x}) = C\left(\frac{f\left(\vec{x}+\vec{l_{\text{in}}}\right)-f(\vec{x})}{\vec{l_{\text{out}}}}\right)
\end{equation}

where $C$ clips the final appended value to zero, and division by $\vec{l_{\text{out}}}$ is elementwise. Let us attempt to differentiate where $f(\vec{x})=n\vec{x}$, an elementwise scaling function:

\begin{equation}
f'(\vec{x}) = C\left(\frac{n\left(\vec{x}+\vec{l_{\text{in}}}\right)-n\vec{x}}{\vec{l_{\text{out}}}}\right)
\end{equation}

\begin{equation}
f'(\vec{x}) = C\left(\frac{n\vec{x}+n\vec{l_{\text{in}}}-n\vec{x}}{\vec{l_{\text{out}}}}\right)
\end{equation}

\begin{equation}
f'(\vec{x}) = C\left(\frac{n\vec{l_{\text{in}}}}{\vec{l_{\text{out}}}}\right)
\end{equation}

Here we strike a problem. Although $\vec{l_{\text{in}}}$ and $\vec{l_{\text{out}}}$ are equal, since $f$ is an elementwise function, dividing one by another is invalid, because all but one of their respective elements are equal to zero, and division of zero by zero is undefined. Regrettably, this form of differentiation is not viable; limits or infinitesimals would need to be re-introduced, defeating the entire object of the orthagonal vector system.

The root cause of this problem is that $\vec{l_{\text{out}}}$ fails to obey the first rule stated above: $\left(\vec{l_{\text{out}}}\right)^0\not=1$. If it had somehow resolved to a vector of ones of length $D_{\text{out}}+1$, the reasoning could have continued as follows:

\begin{equation}
f'(\vec{x}) = C\left(\vec{n}^+\right)
\end{equation}

\begin{equation}
f'(\vec{x}) = \vec{n}
\end{equation}

where the elevated $+$ sign denotes that the vector is in $D_{\text{out}}+1$ dimensions, including the appended value; without it, $\vec{n}$ is the clipped form of only $D_{\text{out}}$ dimensions. This is indeed the result we expect, but it requires doing the impossible in dividing zero by zero.

To round off this enquiry into different forms of differentiation, we look at some Python implementations of differential systems for different functions, both single-input and multivariate.

```python
# We define the imaginary differentiation function of a function f at a point x
def get_imag_deriv_at_point(f,x):
  x = complex(x)
  imag_deriv = (f(x + 1j) + f(x))/1j
  deriv = imag_deriv.real
  return deriv

# Quadratic function (the imaginary differentiation function should work accurately for this)
def square(x):
  return x**2

for x in range(-10,11):
  imag_deriv = get_imag_deriv_at_point(square,x)
  print(f"Imaginary derivative of x^2 at x = {x}: {imag_deriv}. True derivative: {2.0*x}")
```

```python
# Now we'll try a cubic function.
# Given the generalisation problems with cubic functions and higher-order polynomials, we expect it to be off by the
# negative coefficient of the cubic (and only) term, which in this case is 1
def cube(x):
  return x**3

for x in range(-10,11):
  imag_deriv = get_imag_deriv_at_point(cube,x)
  print(f"Imaginary derivative of x^3 at x = {x}: {imag_deriv}. True derivative: {3.0*(x**2)}")
```

```python
# Now we create a general mutlivariate case.
# For simplicity, the functions in this example will take a Numpy array of inputs, rather than individual parameters
# directly.
# Also, we assume all functions output a single scalar.
import numpy as np

def get_imag_deriv_at_points(f,x_vec):
  complex_x_vec = np.array(x_vec,dtype=complex)
  imag_deriv = np.zeros_like(x_vec)
  func_orig = f(complex_x_vec)
  for i, x in enumerate(complex_x_vec):
    complex_x_vec_aug = np.array(complex_x_vec)
    complex_x_vec_aug[i] += 1j
    imag_deriv[i] = (f(complex_x_vec_aug) - func_orig)/1j
  deriv = imag_deriv.real
  return deriv

# We define a multivariate quadratic function, which should also be correctly differentiated
def square_sum(x_vec):
  return np.sum(x_vec**2)

# In this loop, we only find the derivative with respect to the first input to square_sum.
# Changes in the other two values should not effect this, as they are treated as constants in differentiation.
# Try changing them and re-running to see this holds true in practise.
for x_first in range(-10,11):
  x_vec = [x_first,2,4]
  imag_deriv = get_imag_deriv_at_points(square_sum,x_vec)
  print(f"Imaginary derivative of square_sum(x) at x_1 = {x_first}: {imag_deriv[0]}. True derivative: {2.0*x_first}")
```

```python
# We define a multivariate quadratic function.
# We expect the derivative with respect to the first input to be off by one, as before, since the others terms
# are constants, and all coefficients equal 1.
def cube_sum(x_vec):
  return np.sum(x_vec**3)

for x_first in range(-10,11):
  x_vec = [x_first,2,4]
  imag_deriv = get_imag_deriv_at_points(cube_sum,x_vec)
  print(f"Imaginary derivative of cube_sum(x) at x_1 = {x_first}: {imag_deriv[0]}. True derivative: {3.0*(x_first**2)}")
```

```python
# Finally, we will see how the earlier single-variable imaginary derivative function handles f(x) = e^x, which
# was not discussed in the text.
import math

def exp(x):
  return math.e ** x

for x in range(-10,11):
  imag_deriv = get_imag_deriv_at_point(exp,x)
  print(f"Imaginary derivative of e^x at x = {x}: {imag_deriv}. True derivative: {exp(x)}")
```
